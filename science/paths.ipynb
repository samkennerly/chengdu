## POSIX and Windows paths

Windows handles file paths differently from POSIX systems.
`Path` objects should work on POSIX or Windows,
but the differences can be confusing.

### separators

POSIX paths use `/` slashes. Windows uses `\` backslashes.
The `pathlib` library always uses `/`, even for Windows paths.
Using `\` in programs can be very confusing because it's an escape character for strings.

### home folders

These are typical home folders for a user named `dave`:

- Linux: `/usr/home/dave`
- macOS: `/Users/dave`
- Windows: `C:\Users\dave`

### drive letters

Windows paths start with a _drive letter_ like `C:` or `D:`.
I think this is a leftover from when everyone used floppy disks.
My parents' IBM PC had 2 floppy drives named `A:` and `B:`.
Later computers had a hard drive named `C:`.

````python
# When I run this, pathlib creates a PosixPath.
# On a Windows machine, it will be a WindowsPath.
Path.home()
````

````python
# You can do all of these things with strings, but
# I think this is easier, and it works on Windows or POSIX.
path = DELETE_ME
print("This path's parent folder is", path.parent)
print("This path's short name is", path.name)
print("This path's suffix is", path.suffix)
````

## relative paths

An _absolute path_ is like a person's full name.
A _relative path_ is like a person's given name.
For example, if your CWD (Current Working Directory) is:
```
/usr/home/guido/scripts
```
and you want to run this script
```
/usr/home/guido/scripts/chores/backup.py
```
then you can use this relative path:
```
python3 chores/guido.py
```
You don't need to type the absolute path.

````python
# Convert a relative path to an absolute path
relpath = Path("data/hello.json")
abspath = Path.cwd() / relpath
print(f"{relpath} is short for {abspath}")
````

## inspect the data folder

**BONUS PROBLEM FOR BORED PEOPLE:**
Look up `glob` syntax for finding paths recursively. Many languages and operating systems use `glob` syntax.

````python
# Does the data folder exist? Is it a folder (directory)?
DATA.exists(), DATA.is_dir()
````

````python
# What files are in the data folder?
files = [x for x in DATA.iterdir() if x.is_file()]

# Filter out dotfiles. I don't want to see them.
files = [x for x in files if not x.name.startswith('.')]

# Show file paths in order.
sorted(files)
````

````python
# What folders are in the data folder?
sorted(x for x in DATA.iterdir() if x.is_dir())
````

````python
# What's in the test folder?
sorted((DATA / "test").iterdir())
````

````python
# 'glob' is an old UNIX program for finding files.
# Path.glob() uses the same syntax as glob.
sorted(DATA.glob("**/*.csv"))
````

## save a file

````python
# This code works with Path objects or strings.
with open(DELETE_ME, 'w') as file:
    file.write("Please delete this file.")

# For text-like files, Path objects can also do this.
DELETE_ME.write_text("Please delete this file.")
````

## read a file

````python
# This code works with Path objects or strings.
with open(DELETE_ME) as file:
    print(file.read())

# For text-like files, Path objects can also do this.
print(DELETE_ME.read_text())
````

## delete a file
Before you delete _anything_, be sure it's something you want to delete.

````python
DELETE_ME.unlink()
DELETE_ME.exists()
````
