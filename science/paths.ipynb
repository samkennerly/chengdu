# file paths
To read or save files, you need to find them.

```python
from pathlib import Path
```

`pathlib` is a Python package of file path tools.
It's builtin, so it doesn't need pip, venv, conda, etc.

I rarely use the whole `pathlib` package. I just use `Path` objects.

```python
# ALL_CAPS names mean "We might need this. Don't mutate it!"
HERE = Path.cwd()
REPO = HERE.parent
DATA = HERE / "data"
DELETE_ME = DATA / "delete_me.txt"

print("The current working directory is", HERE)
print("This repository is", REPO)
print("Datasets are in", DATA) 
print("Use this path for practice:", DELETE_ME)
```

I use a POSIX OS, so my file paths all have `/` slashes. Windows file paths have `\` backslashes.

`Path` objects should work on POSIX or Windows, but there are some weird exceptions.

```python
# Path objects are easy to cast as strings.
str(REPO)
```

## What's in the data folder?

```python
# Does the data folder exist? Is it a folder (directory)?
DATA.exists(), DATA.is_dir()
```

```python
# What files are in the data folder?
files = [x for x in DATA.iterdir() if x.is_file()]

# Filter out dotfiles. I don't want to see them.
files = [x for x in files if not x.name.startswith('.')]

# Show file paths in order.
sorted(files)
```

```python
# What folders are in the data folder?
sorted(x for x in DATA.iterdir() if x.is_dir())
```

```python
# What's in the test folder?
sorted((DATA / "test").iterdir())
```

```python
# 'glob' is an old UNIX program for finding files.
# Path.glob() uses the same syntax as glob.
sorted(DATA.glob("**/*.csv"))
```

## save a file

```python
# This code works with Path objects or strings.
with open(DELETE_ME, 'w') as file:
    file.write("Please delete this file.")

# This is how I usually write text-like files.
DELETE_ME.write_text("Please delete this file.")
```

## read a file

```python
# This code works with Path objects or strings.
with open(DELETE_ME) as file:
    print(file.read())

# This is how I usually read text-like files.
print(DELETE_ME.read_text())
```

## delete a file

```python
DELETE_ME.unlink()
DELETE_ME.exists()
```
