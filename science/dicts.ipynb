# dictionaries and json files

The `dict()` function creates dictionaries.

Dictionaries are Python's version of a _key-value store_. Many languages have key-value stores, and _NoSQL_ databases (like Redis and Mongo) are often used as key-value stores. A key-value store is a bit like a function, but with hard-coded inputs (keys) and outputs (values).

```python
from pathlib import Path
import json
```

```python
HELLO_PATH = Path.cwd() / "data/hello.json"
```

## create a dictionary

```python
# There are many ways to create a dictionary.
# This way is popular. It looks like a JSON file.
hello = {
    "morning": "早上好",
    "afternoon": "下午好",
    "evening": "晚上好",  # This , is optional.
}
hello
```

```python
# Dictionary keys are a special type.
print(hello.keys())

"morning" in hello.keys()
```

```python
# Dictionary values are another special type.
print(hello.values())

"你好" in hello.values()
```

## iterate over a dictionary

```python
# This is an infamous (famous for bad reasons) trap.
# If we try to cast a dictionary to list (or tuple, set,etc.),
# the dictionary's values are ignored. We only get keys.
list(hello)
```

```python
# If we iterate (loop) over a dictionary, we only see keys.
[x.upper() for x in hello]
```

```python
# Use .items() to loop over keys and values together.
for key, val in hello.items():
    print(f"In the {key}, say {val}")
```

```python
# Iterating over .items() generates tuples.
# (So do zip() and enumerate().)
[(key.upper(), val) for key, val in hello.items()]
```

```python
# This returns a list of (key, value) tuples.
list(hello.items())
```

```python
# We can also create dictionaries by looping over tuples.
names = "Ada Charles Grace Guido Linus Sam".split()
scores = [100, 90, 94, 86, 0, 42]
dict(zip(names, scores))
```

## get values

The main purpose of a key-value store is: given a key, return its value quickly.

```python
# Get the value for key "morning"
hello["morning"]
```

```python
# If you input a key that is not in the dictionary,
# Python will raise a KeyError.
try:
    hello["lunch"]
except KeyError as err:
    print("Error!", err)
```

```python
# The .get() method does not raise a KeyError.
# If you input an unknown key, it will output None.
print(hello.get("lunch"))
```

```python
# .get() can also return a (defval) default value
hello.get("lunch", "你好")
```

## set values

Dictionaries are mutable. We can add keys, delete keys, and change values.

Dictionaries do _not_ allow duplicate keys. If we try to add a key that already exists, the old value will be overwritten.

**BONUS PROJECT FOR BORED PEOPLE**: Python has `tuple()` to make immutable lists and `frozenset()` to make immutable sets,
but it has no builtin immutable dictionary. Make your own! Call it `class FrozenDict`.

```python
# Add a new key and set its value.
hello["lunch"] = "午安"

# Change the value for a key
hello["morning"] = "早啊"

# Show what's in the dictionary now
for key, value in hello.items():
    print(f"When it is {key}, I say {value}.")
```

## JSON strings and files

JSON (JavaScript Object Notation) strings are a standard way to
persist objects in the JavaScript language.

JavaScript _serializes_ an object by converting it to a string that follows JSON rules. That string can be sent to other programs or saved to a file. The `.json` file suffix means "This file is plain text that follows the JSON protocol."

JSON is important to data scientists and engineers because many APIs (Application Programming Interfaces) use JSON strings as inputs and outputs.

```python
# json.dumps() converts a dict (or list) to a JSON string.
# The 'indent = 2' kwarg is not required.
# It just makes the string easier to read.
jstr = json.dumps(hello, indent=2)
print(jstr)

# json.loads() converts a JSON string to a dict (or list).
json.loads(jstr) == hello
```

```python
# json.dump() saves a dict (or list) to a .json file.
# We could also use json.dumps() to make a string,
# then save the string with Path.write_text().
print(f"Save JSON data to {HELLO_PATH}")
with open(HELLO_PATH, 'w') as file:
    json.dump(hello, file, indent=2)

# json.load() reads a dict (or list) from a .json file.
# We could also use Path.read_text() to read the file,
# then convert the string with json.loads().
print(f"Read JSON data from {HELLO_PATH}")
json.load(HELLO_PATH.open()) == hello
```
