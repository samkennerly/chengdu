# dictionaries and json files

The `dict()` function creates dictionaries.
Python dictionaries and many _NoSQL_ databases (like Redis and Mongo) are _key-value stores_. It's a bit like a function, but with hard-coded inputs (keys) and outputs (values).

## create a dictionary

```python
hello = {
    "morning": "早上好",
    "afternoon": "下午好",
    "evening": "晚上好",
}
hello
```

```python
# Dictionary keys are a special type.
print(hello.keys())

"morning" in hello.keys()
```

```python
# Dictionary values are another special type.
print(hello.values())

"你好" in hello.values()
```

## iterate over a dictionary

```python
# If we try to cast a dictionary to list (or tuple or set),
# we only get the keys. The values are ignored.
list(hello)
```

```python
# If we iterate (loop) over a dictionary, we only see keys.
[x.upper() for x in hello]
```

```python
# .items() is used to loop over keys and values together.
for key, val in hello.items():
    print(f"In the {key}, say {val}")
```

```python
# Iterating over .items() generates tuples.
# (So do to zip() and enumerate().)
[(key.upper(), val) for key, val in hello.items()]
```

```python
# This returns a list of (key, value) tuples.
list(hello.items())
```

```python
# We can also create dictionaries by looping over tuples.
names = "Ada Charles Grace Guido Linus Sam".split()
scores = [100, 90, 94, 86, 0, 42]
dict(zip(names, scores))
```

## get values

The main purpose of a dictionary is: given a key, return its value.

```python
# Get the value for key "morning"
hello["morning"]
```

```python
# If you input a key that is not in the dictionary,
# Python will raise a KeyError.
try:
    hello["lunch"]
except KeyError as err:
    print("Error!", err)
```

```python
# The .get() method does not raise a KeyError.
# If you input an unknown key, it will output None.
hello.get("lunch") is None
```

```python
# .get() can also return a default value
hello.get("lunch", "你好")
```

## set values

Dictionaries are mutable. We can add keys, delete keys, and change the value for a key. Dictionaries do _not_ allow duplicate keys. If we try to add a key that already exists,
the old value will be overwritten.

```python
# Add a new key and set its value.
hello["lunch"] = "午安"

# Change the value for a key
hello["morning"] = "早啊"

for key, value in hello.items():
    print(f"When it is {key}, I say {value}.")
```

## JSON strings and files

JSON (JavaScript Object Notation) strings are a standard way to
persist JavaScript objects. An object is _serialized_ by
converting it to a string that follows JSON rules.
That string can be sent to other computers or saved to a file.The `.json` file suffix means "This file is plain text that follows the JSON protocol."

JSON is very important to data scientists and engineers because many APIs (Application Programming Interface) use JSON strings as inputs and outputs.
