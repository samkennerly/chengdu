## escape characters

```python
# If you need a quote mark in a string,
# this is one way to do it...
print("Sam's club")

# or we can use \ as an escape character.
print("Sam\'s club")
print('Sam\'s club')
```

```python
# \ is used for some special characters.
# \n means "start a new line".
# \t means "press the tab key".
print("我吃了\n冰箱\n里的\n李子")
```

## strings are iterable

```python
# Strings have indexes, slices, loops, and 'in'.
# They're a lot like lists of Unicode characters.
hello = "Hello, World!"
print(hello[0:6])
print(hello[6:-1])
print(hello[-1])

vowels = "aeiou"
[x for x in hello if x in vowels]
```

## f-strings

```python
# I use f-strings when I want to print some text,
# and part of that text is a variable.
# This is also called "string interpolation" or "templating".
# It's very important for front-end web developers.
price = 15
text = f"""
This 酸辣面 costs {price} RMB.
That's roughly {round(price / 7)} USD.
"""
print(text)
```

```python
# f-strings have special rules for number formatting.
# These rules are ugly and complicated.
# I often forget them and look them up.
price = 100_000_000
text = f"""
This 飞机 costs {price:,.0f} RMB.
That's roughly {price / 7:,.2f} USD.
"""
print(text)
```

## .format() and `%s`

```python
# Older versions of Python had no f-strings.
# These are other ways to format strings.
price = 150_000
print("This 电动车 costs {} RMB.".format(price))
print("That's roughly %s USD." % round(price / 7))
```

## .join()

```python
# .join() joins a list of strings into one big string.
" & ".join(["Ada", "Charles", "Linus"])
```

```python
# "".join() uses a blank string as a separator.
# It concatenates, just like the + operator.
"".join(["Ada", "Charles", "Linus"])
```

## .split()

```python
# When .split sees whitespace (like space, tab, or newline),
# it partitions a string and returns a list of parts.
"Hello, World!".split()
```

```python
# .split() can use other characters as separators
"automobile".split("o")
```

```python
# Using \n as separator splits a string into lines.
# It preserves other whitespace like tabs and spaces.
DAO_PATH.read_text().split("\n")
```

## more string methods

For even more string methods, see the official docs:

https://docs.python.org/3/library/string.html

```python
# Strings have many special methods.
# These are the methods I use most often.
hello = "Hello, World!"
print(hello.lower())
print(hello.upper())
print(hello.strip("!"))
print(hello.startswith("Hello"))
print(hello.endswith("!"))
print(hello.replace('!', '?'))
```

```python
# "Case-folding" avoids a common data science trap.
# If I'm reading a list of students, my name might be
# 'sam' in some places and "Sam" in others.
# I avoid this by forcing all strings to lowercase.
students = ["Ada", "Charles", "Sam"]
print("Is sam here?", "sam" in students)

"sam" in [x.lower() for x in students]
```

```python
# "Stripping whitespace" avoids another common trap.
# When people use spreadsheets (like Excel), they often
# add extra spaces or tabs to cells. They can't see
# whitespace, so they don't notice the problem.
students = ["Ada  ", "Charles   ", "\tSam        "]
print("Is Sam here?", "Sam" in students)

"sam" in [x.strip().lower() for x in students]
```
