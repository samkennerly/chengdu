# strings and txt files

The `.txt` suffix means "This file is plain text."

Plain text files are standard way to _persist_ strings in a format that humans and (almost) all computers can read.

## Python quote marks

```python
# "double quotes" mean "This is not code. It's a string."
nihao = "你好"

# 'single quotes' mean the same thing.
zaijian = "再见"

nihao, zaijian
```

```python
# '''triple quotes''' can be multiple lines
poem = '''
I have eaten
the plums
that were in
the icebox
'''
print(poem)

# """triple double quotes""" can also be multiple lines
poem = """
VSCode has deleted
the code
that was in
my notebook
"""
print(poem)
```

## escape characters

```python
# If you need a quote mark in a string,
# this is one way to do it...
print("Sam's club")

# Or we can use \ as an escape character.
print("Sam\'s club")
print('Sam\'s club')
```

```python
# \ is also used for some special characters.
# \n means "start new line".
# \t means "tab character".
print("我吃了\n冰箱\n里的\n李子")
```

## strings are iterable

```python
# Like lists, strings have ndexes, slices, loops, and 'in'
hello = "Hello, World!"
print(hello[0:6])
print(hello[6:-1])
print(hello[-1])

vowels = "aeiou"
[x for x in hello if x in vowels]
```

## f-strings

```python
# I use f-strings when I want to print some text,
# and part of that text is a variable.
price = 15
text = f"""
This 酸辣面 costs {price} RMB.
That's roughly {round(price / 7)} USD.
"""
print(text)
```

```python
# f-strings have special rules for number formats
price = 100_000_000
text = f"""
This car costs {price:,.0f} RMB.
That's roughly {price / 7:,.2f} USD.
"""
print(text)
```

## .format() and `%s`

```python
# These are other ways to format strings.
price = 15
print("This 酸辣面 costs {} RMB.".format(price))
print(f"That's roughly %s USD." % round(price / 7))
```

## .join()

```python
# .join() joins a list of strings into one big string.
# This example concatenates (like the + operator):
" & ".join(["Ada", "Charles", "Linus"])
```

```python
# "".join() uses a blank string as a separator.
# It concatenates, just like the + operator.
"".join(["Ada", "Charles", "Linus"])
```

## .split()

```python
# When .split sees whitespace (like space, tab, or newline),
# it partitions a string and returns a list of parts.
"Hello, World!".split()
```

```python
# .split() can use other characters as separators
"automobile".split("o")
```

```python
# Using \n as separator splits a string into lines.
DAO_PATH.read_text().split("\n")
```

## some more string methods

```python
hello = "Hello, World!"
print(hello.lower())
print(hello.upper())
print(hello.strip("!"))
print(hello.startswith("Hello"))
print(hello.endswith("!"))
```
