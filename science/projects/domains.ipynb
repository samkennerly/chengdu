# domains

Each row is a TLD (top-level domain) for websites. Columns are:
- `Domain` the TLD
- `Type` Is it generic, country-code, or something else?
- `Sponsoring Organization` Who is responsible for this TLD?


## project ideas

- Count domains for each type.
- Find all domains that start with the letter `u`.
- Find all domains that start with the character `中`.


```python
from pathlib import Path
import pandas as pd
```

```python
# Where is the dataset on your computer?
DATA_PATH = Path.cwd().parent / "data" / "domains.csv"
print(f"The dataset is here: {DATA_PATH}")
```

```python
def cleanstr(values):
    # Convert strings to lowercase without whitespace.
    return [x.lower().replace(' ', '_').strip() for x in values]
```

```python
# Read raw data.
domains = pd.read_csv(DATA_PATH)
print(f"{len(domains)} rows loaded")

# Normalize column names.
domains.columns = cleanstr(domains.columns)

# Each domain is already a unique ID.
domains = domains.set_index('domain').sort_index()

domains
```

```python
# Check data types
domains.dtypes
```

```python
# Count null markers in each column
domains.isnull().sum()
```

```python
# Show all domain types
for x in sorted(set(domains['type'])):
    print(x)
```

```python
# Example project: Find all non-ASCII domains.
# ASCII is an old standard for encoding characters.
# It includes these letters in lower and uppercase:

from string import ascii_lowercase

print("ASCII lowercase letters are", ascii_lowercase)

def is_ascii(tld):
    return all(x.lower() in ascii_lowercase for x in tld)

tlds = [x.strip('.') for x in domains.index]
tlds = [x for x in tlds if not is_ascii(x)]
print(f"{len(tlds)} non-ASCII TLDs found")

tlds
```
